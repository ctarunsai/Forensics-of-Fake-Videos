```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
import json
```

```python
file = open('metadata.json')
```

```python
data = json.load(file)
```

```python
print(data)
```

```python
count_fake = 0
count_real = 0
```

```python
for i in data.values():
    if(i['label']=='FAKE'):
        count_fake += 1
    else:
        count_real += 1
```

```python
print('Real image count is : ',count_real)
print('Fake image count is : ',count_fake)
```

```python
labels = 'REAL','FAKE'
sizes = [count_real,count_fake]
colors = ['gold','lightcoral']
explode = [0.04,0.04]
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.pie(sizes,explode=explode,colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
